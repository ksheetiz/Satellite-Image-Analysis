```python
class_count = pd.DataFrame(df.sum(axis=0)).reset_index()
class_count.columns = ["class", "Count"]
class_count.drop(class_count.index[0], inplace=True)
```

```python
# The Following Visualisation Indicates the class imbalances in the dataset
fig, ax= plt.subplots(figsize=(12,10))
sb.barplot(y="class", x="Count",  data=class_count, ax=ax);
```
